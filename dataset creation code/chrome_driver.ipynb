```python
def getting_text(driver, soup1):
    try:
        material = soup1.find('article') 
        if material is not None:
            real_text = []
            for abzats in material.find_all('p'):
                if abzats is not None:
                    real_text.append(abzats.get_text())
                    real_text.append('  ')
            real_text1 = ''.join(real_text)
            return real_text1
    except:
        return 'N/A'
    
# Как возвращать все абзацы    
```

```python
def to_comment(para, host, driver):
    comment = para.next_sibling
    driver.get(host+comment.find('a').get('href'))
    sleep(1)
                
    page_com = requests.get(driver.current_url)
    soup_com = bs(page_com.text, 'lxml')
    komm = soup_com.find('article')
                
    if komm != None:
        ctext = []
        divs = komm.find('div')
        if divs is True:
            for div in komm.find_all('div'):
                p1 = div.find('p')
                if p1 is True:
                    for part in komm.find_all('p'):
                        if part is not None:
                            ctext.append(part.get_text())
                            ctext.append('  ')
                else:
                    ctext.append(div.get_text())
                    ctext.append('  ')
                    
        else:
            for part in komm.find_all('p'):
                if part is not None:
                    ctext.append(part.get_text())
                    ctext.append('  ')
        ctext1 = ''.join(ctext)
        return ctext1
    else:
        return 'N/A'
    
# Как возвращать все абзацы
```

```python
def getting_comment(driver, host):
    sleep(1)
    body = soup1.find_all('div', class_='b-feed__body')[1]
    try:
        para = body.find('h2', text = re.compile('Комментарии Российской Газеты'), 
                                                    class_='b-feed__name')
        if para != None:
            return to_comment(para, host, driver)

    except IndexError:
        return 'N/A'
```

```python
def document_date(soup1):
    date = soup1.find('span', class_='b-material-head__date-day')
    if date is not None:
        return date.get_text()
    else:
        return 'N/A'
```
