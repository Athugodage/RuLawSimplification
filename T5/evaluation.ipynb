```python
from transformers import pipeline

model = pipeline('text2text-generation', model='marcus2000/T5-RLS500')


## https://huggingface.co/marcus2000/legal_text_simplifier - without NLI
```

```python
from datasets import load_dataset

dataset = load_dataset('marcus2000/twowaydata')
```

```python
predicted = model(dataset['test']['0'][:10])

```

```python
from tqdm import tqdm


predicted = []

for text in tqdm(dataset['test']['0']):
    simplified = model(text)[0]
    predicted.append(simplified['generated_text'])
```

```python
import evaluate

sari = evaluate.load("sari")
sari_score = sari.compute(sources=list(dataset['test']['0'][:500]), 
                          predictions=[pred for pred in predicted[:500]], 
                          references=[[ref] for ref in list(dataset['test']['1'][:500])])

print(sari_score)
```
